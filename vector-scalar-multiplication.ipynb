```python
# import modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
vec = np.array([3,4,5,2])
s = 3

vec*s
```

```python
vec2d = np.array([1,2])
s1 = 2
s2 = .5
s3 = -1

# indexing
vec2d[0]
```

```python
plt.plot([0,vec2d[0]],[0,vec2d[1]], 'bs-',label='v')
plt.plot([0,s1*vec2d[0]],[0,s1*vec2d[1]], 'ro-', label='v*s1')
plt.plot([0,s2*vec2d[0]],[0,s2*vec2d[1]], 'kp-', label='v*s2')
plt.plot([0,s3*vec2d[0]],[0,s3*vec2d[1]], 'g*-', label='v*s3')

plt.axis('square')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.grid()
plt.legend()
plt.show()
```

```python
# assignment!
# create two vectors (2D vectors)
vec2dfirst = np.array([5,2])
vec2dsecond = np.array([7,1])
# plot each vector
plt.plot(vec2dfirst, 'kp-', label='vec2dfirst')
plt.plot(vec2dsecond,'bs-',label='vec2dsecond')
# plot vector v1+v2, v1-v2, (v1*4 + v2/2)
plt.plot([0,vec2dfirst[0]+vec2dsecond[0]],[0,vec2dfirst[1]+vec2dsecond[1]], 'g*-', label='vec2dfirst+vec2dsecond')
plt.plot([0,vec2dfirst[0]-vec2dsecond[0]],[0,vec2dfirst[1]-vec2dsecond[1]], 'kp-', label='vec2dfirst-vec2dsecond')
plt.plot([0,vec2dfirst[0]*4+vec2dsecond[0]/2],[0,vec2dfirst[1]*4+vec2dsecond[1]/2], 'ro-', label='vec2dfirst*4 + vec2dsecond/2')
plt.grid()
plt.legend()
plt.show()
```
