```python

# Coordenadas de las dos rectas
x1, y1 = [0, 3], [3, 0]  # línea 1
x2, y2 = [0, 3], [0, 3]  # línea 2

# Coeficientes de las ecuaciones de las rectas
m1, b1 = np.polyfit(x1, y1, 1)
m2, b2 = np.polyfit(x2, y2, 1)

# Intersección
xi = (b2 - b1) / (m1 - m2)
yi = m1 * xi + b1

print(f"Punto de intersección: ({xi}, {yi})")
```
