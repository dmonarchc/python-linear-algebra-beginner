```python
#import modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
v1 = np.array([5,6,2])
v2 = np.array([3,0,5])

np.dot(v1,v2)
```

```python
v3 = np.array([-4,3,1])
np.dot(v2,v3)
```

```python
v4 = np.array([0,0,0])
np.dot(v3,v4)
```

```python
v5 = np.array([2,3,-1])
np.dot(v3,v5)
# vector v3 is orthogonal to vector v5
```

```python
# assignment
# 1) come up with 3 2D vectors, two of them are orthogonal, but neither not orthogonal to the third
vector1 = np.array([6,4])
vector2 = np.array([-2,3])
vector3 = np.array([3,7])
# 2) plot all three vectors --make sure the axes are square and equal
plt.plot(vector1,label='vector1')
plt.plot(vector2,label='vector2')
plt.plot(vector3,label='vector3')
plt.legend()
plt.show()
```

```python

# aqui se ve mejor que pasa con dos vectores ortogonales
vector1 = np.array([6,4])
vector2 = np.array([-2,3])
vector3 = np.array([3,7])

# Graficar vectores desde el origen
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label="vector1")
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='g', label="vector2")
plt.quiver(0, 0, vector3[0], vector3[1], angles='xy', scale_units='xy', scale=1, color='b', label="vector3")

# Configuracin de ejes
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.gca().set_aspect('equal', adjustable='box')  # ejes cuadrados e iguales
plt.axhline(0, color='black', linewidth=0.5)  # eje x
plt.axvline(0, color='black', linewidth=0.5)  # eje y
plt.legend()
plt.grid(True)
plt.show()
```
